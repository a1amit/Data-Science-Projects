## Question 1: Part 2

```python
IRONMAN_NAME = "Iron Man/TONY STARK"
CAPTAIN_AMERICA_NAME = "Captain America"
SPIDERMAN_NAME = "SPIDER-MAN/PETER PAR"


bigThree = df.loc[df['hero2'].str.contains('Iron Man/TONY STARK|Captain America|Spider-Man/Peter Par', case =False)]

ironman_friends = list()
captain_america_friends = list()
spiderman_friends = list()

#creating alists of the friends for each of the three heros(the lists takes into account hero 1 and hero 2 column)
for index, row in bigThree.iterrows():
  hero1, hero2 = row['hero1'], row['hero2']
  if hero1.lower().strip() == IRONMAN_NAME.lower().strip():
      ironman_friends.append(hero2)
  elif hero2.lower().strip() == IRONMAN_NAME.lower().strip():
      ironman_friends.append(hero1)

  if hero1.lower().strip() == CAPTAIN_AMERICA_NAME.lower().strip():
      captain_america_friends.append(hero2)
  elif hero2.lower().strip() == CAPTAIN_AMERICA_NAME.lower().strip():
      captain_america_friends.append(hero1)
      
  if hero1.lower().strip() == SPIDERMAN_NAME.lower().strip():
      spiderman_friends.append(hero2)
  elif hero2.lower().strip() == SPIDERMAN_NAME.lower().strip():
      spiderman_friends.append(hero1)


# print(list(set(ironman_friends)))
# print(list(set(captain_america_friends)))
# print(list(set(spiderman_friends)))

NUMBER_OF_FRIENDS =0
for friend in set(ironman_friends + captain_america_friends + spiderman_friends):
  if (friend in captain_america_friends and friend in ironman_friends and friend in spiderman_friends):
    NUMBER_OF_FRIENDS += 1

print(f"Number Of Common friends: {NUMBER_OF_FRIENDS}")
```

**Bigger than 5:**

```python
df_biggerThan5 = df.loc[(df['Magnitude'] > 5)]
df_biggerThan5['Year'] = df_biggerThan5['Date'].str.slice(start=6)
df_biggerThan5 = df_biggerThan5.groupby(['Year']).count()
df_biggerThan5 = df_biggerThan5.sort_values(by = 'Date', ascending= False).head(1)

print("the year with the max number of earthquakes above the magnitude of 5:  ")
df_biggerThan5
```
